```python
import json
#people u may know
data=json.load(open("data.json","r+"))
def people_u_may_know(u_id,data):
    user_friend={}
    for user in data['users']:
        user_friend[user["id"]]=set(user['friends'])
    if u_id not in user_friend:
        return []
    suggestion=[]
    for friend in user_friend:
        for mutual in user_friend[friend]:
            if mutual!=u_id and mutual not in user_friend[u_id]:
                suggestion.append(mutual)
    return suggestion
u_id=int(input("entre the user id "))
mutual=people_u_may_know(u_id,data)
print(mutual)
```
