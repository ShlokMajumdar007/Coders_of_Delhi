```python
import json
data=json.load(open("data.json","r"))
def clean(data):
    #removal of empty user
    data['users']=[users for users in data['users'] if users['name'].strip()]
    #removal of dublicate friends
    for user in data['users']:
        user['friends']=list(set(user['friends']))
    #removal of inactive user
    data["users"] = [user for user in data["users"] if user["friends"] or user["liked_pages"]]
    return data
cleaned_data=clean(data)
json.dump(cleaned_data,open("cleaned_data.json","w+"))
print("done")
```

```python
#people u may know 
import json
data=json.load(open("cleaned_data.json","r+"))
def people_u_may_know(user_id,data):
    friends={}
    for user in data['users']:
            friends[user['id']]=set(user['friends'])
    if user_id not in friends:
        return []
    suggestion=[]
    direct_friends=list(friends[user_id])
    for friend in direct_friends:
        for mutual in friends[friend]:
            if mutual != user_id and mutual not in direct_friends:
                suggestion.append(mutual)
    return set(suggestion)

user_id=int(input("entre the user_id:"))
people=people_u_may_know(user_id,data)
print("the people ",user_id,"may know",people)
```
